```python
import fitz  # PyMuPDF
import re
import joblib

SECTION_HEADERS = [
    "summary", "objective", "profile", "education", "qualifications", "experience",
    "work experience", "projects", "skills", "technical skills",
    "certifications", "achievements", "contact", "personal details"
]

def is_section_header(line):
    norm = line.lower().strip(":").strip()
    return any(norm.startswith(h) for h in SECTION_HEADERS)

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    full_text = ""
    for page in doc:
        full_text += page.get_text()
    return full_text

def split_into_sections(text):
    sections = {}
    current = "general"
    sections[current] = []

    lines = text.split("\n")
    for line in lines:
        clean = line.strip()
        if not clean:
            continue

        if is_section_header(clean):
            current = clean.lower().strip(":").strip()
            if current not in sections:
                sections[current] = []
        else:
            sections.setdefault(current, []).append(clean)

    return sections

def extract_features(token, index):
    return {
        'token': token.lower(),
        'is_capitalized': token[0].isupper(),
        "is_upper": token.isupper(),
        'is_all_caps': token.isupper(),
        'is_title': token.istitle(),
        'prefix-1': token[:1],
        'prefix-2': token[:2],
        'suffix-1': token[-1:],
        'suffix-2': token[-2:]
    }

def predict_names(tokens):
    clf = joblib.load("name_model.joblib")       # Make sure this model supports predict_proba()
    vec = joblib.load("name_vectorizer.joblib")

    features = [extract_features(tok, idx) for idx, tok in enumerate(tokens)]
    X_vec = vec.transform(features)
    probs = clf.predict_proba(X_vec)  # Get probabilities for each class

    # Class 1 is usually the "name" label
    for tok, prob in zip(tokens, probs):
        print(f"{tok:<15} â†’ NAME prob: {prob[1]:.2f} (Non-name: {prob[0]:.2f})")

    # Get the most confident name predictions
    name_tokens = [tok for tok, prob in zip(tokens, probs) if prob[1] > 0.5]
    return " ".join(name_tokens)
```

```python
pdf_path = "Charitha Resume.pdf"  # path to resume
text = extract_text_from_pdf(pdf_path)

sections = split_into_sections(text)
print(sections)

for sec, lines in sections.items():
    print(f"\n== {sec.upper()} ==")
    for l in lines:
        tokens=l.split()
        print("Predicted Name:", predict_names(tokens))
```

```python
import fitz  # PyMuPDF

def extract_pdf_text(filepath):
    text = ""
    with fitz.open(filepath) as doc:
        for page in doc:
            text += page.get_text()
    return text


import re

def split_resume_sections(text):
    sections = {}
    current = "general"
    sections[current] = []

    lines = text.split("\n")
    for line in lines:
        clean = line.strip()
        if not clean:
            continue

        if re.match(r"^[A-Z\s&]+$", clean) and len(clean.split()) <= 5:
            current = clean.strip().lower()
            sections[current] = []
        else:
            sections.setdefault(current, []).append(clean)

    if not sections["general"]:
        del sections["general"]
    return sections
```

```python
import fitz  # PyMuPDF
import re

def extract_pdf_text(filepath):
    text = ""
    with fitz.open(filepath) as doc:
        for page in doc:
            text += page.get_text()
    return text

def extract_resume_header(text):
    lines = text.splitlines()
    header_lines = []

    # Define what counts as the first "heading"
    heading_pattern = re.compile(r"^[A-Z\s&]{4,}$")  # e.g., SUMMARY, TECHNICAL SKILLS

    for line in lines:
        stripped = line.strip()
        if heading_pattern.match(stripped):
            break  # Stop at the first detected heading
        if stripped:  # Skip empty lines
            header_lines.append(stripped)

    return header_lines

# Usage
pdf_path = "Charitha_dokku_resume.pdf"
full_text = extract_pdf_text(pdf_path)
header = extract_resume_header(full_text)

# Print header content
print("=== HEADER ===")
for line in header:
    print(line)
```

```python
import fitz  # pip install pymupdf

def extract_page_header_fitz(page, size_diff=2):
    blocks = page.get_text("dict")["blocks"]
    spans = [span for block in blocks for span in block.get("spans", [])]

    if not spans:
        return ""  # No text on page, or no spans found

    max_font = max(span["size"] for span in spans)
    header_lines = []

    for span in spans:
        text = span["text"].strip()
        size = span["size"]
        if not text:
            continue
        # First heading detected
        if size >= max_font - size_diff:
            break
        header_lines.append(text)

    return "\n".join(header_lines).strip()


def extract_headers_from_pdf(path):
    doc = fitz.open(path)
    results = []
    for page in doc:
        header = extract_page_header_fitz(page)
        results.append(header)
    return results

# Run it
headers = extract_headers_from_pdf("Charitha_dokku_resume.pdf")
for i, h in enumerate(headers, 1):
    print(f"\nâ€” Page {i} HEADER â€”\n{h or '[No header found]'}\n")
```

```python
import re
import fitz  # pip install pymupdf

SECTION_LABELS = [
    "summary", "objective", "education", "skills", "technical skills",
    "projects", "experience", "certifications", "achievements",
    "internships", "contact", "additional information"
]

def extract_text_from_pdf(path):
    doc = fitz.open(path)
    return "\n".join(page.get_text() for page in doc)

def normalize_line(line):
    return re.sub(r"[^a-zA-Z\s&]", "", line).lower().strip()

def split_sections_generic(text):
    sections = {}
    current = "general"
    sections[current] = []

    lines = text.splitlines()
    for line in lines:
        clean = line.strip()
        if not clean:
            continue

        norm_line = normalize_line(clean)
        if norm_line in SECTION_LABELS:
            current = norm_line
            sections[current] = []
        else:
            sections.setdefault(current, []).append(clean)

    if not sections["general"]:
        del sections["general"]
    return sections

# ðŸ”½ Run this to see your section-wise output
text = extract_text_from_pdf("Charitha Resume.pdf")
print(f"text is {text}")
sections = split_sections_generic(text)


```

```python
import pdfplumber
import matplotlib.pyplot as plt

def extract_and_visualize_sections(pdf_path, spacing_threshold=20):
    with pdfplumber.open(pdf_path) as pdf:
        page = pdf.pages[0]  # Process only the first page for now
        words = page.extract_words(use_text_flow=True, keep_blank_chars=True)

        # Sort by vertical position, then by horizontal
        words.sort(key=lambda x: (x['top'], x['x0']))

        # Group words into paragraphs/sections based on vertical gap
        paragraphs = []
        current_paragraph = []
        prev_bottom = None

        for word in words:
            top = word['top']
            bottom = word['bottom']

            if prev_bottom is not None and abs(top - prev_bottom) > spacing_threshold:
                if current_paragraph:
                    paragraphs.append(current_paragraph)
                    current_paragraph = []
            current_paragraph.append(word)
            prev_bottom = bottom

        if current_paragraph:
            paragraphs.append(current_paragraph)

        # Print section contents
        print("\n--- SECTION CONTENTS ---\n")
        for i, para in enumerate(paragraphs, 1):
            text = " ".join(word['text'] for word in para)
            print(f"[Section {i}]\n{text}\n")

        # Visualization
        fig, ax = plt.subplots(figsize=(8, 11))
        ax.set_xlim(0, page.width)
        ax.set_ylim(page.height, 0)
        ax.set_title("Detected Sections on PDF Page")

        for i, para in enumerate(paragraphs):
            x0 = min(word['x0'] for word in para)
            x1 = max(word['x1'] for word in para)
            top = min(word['top'] for word in para)
            bottom = max(word['bottom'] for word in para)

            rect = plt.Rectangle((x0, top), x1 - x0, bottom - top, edgecolor='red', facecolor='none', lw=2)
            ax.add_patch(rect)
            ax.text(x0, top - 5, f"Section {i+1}", color='blue', fontsize=8)

        plt.gca().invert_yaxis()
        plt.tight_layout()
        plt.show()

# Example usage:
extract_and_visualize_sections("Mayukha Resume.pdf")
```

```python
import fitz  # PyMuPDF
import matplotlib.pyplot as plt

def extract_sections_with_fitz(pdf_path, spacing_threshold=20):
    doc = fitz.open(pdf_path)
    page = doc[0]  # First page

    # Extract all text blocks [(x0, y0, x1, y1, "text", block_no, block_type)]
    blocks = page.get_text("blocks")
    blocks = [b for b in blocks if b[4].strip()]  # remove empty text blocks

    # Sort blocks by y0 (top coordinate in PDF space is higher y)
    blocks.sort(key=lambda b: b[1])  # sort top-down

    # Group into sections by vertical gap
    sections = []
    current_section = []
    prev_y1 = None

    for block in blocks:
        x0, y0, x1, y1, text, *_ = block
        if prev_y1 is not None and (y0 - prev_y1) > spacing_threshold:
            if current_section:
                sections.append(current_section)
                current_section = []
        current_section.append(block)
        prev_y1 = y1

    if current_section:
        sections.append(current_section)

    # Print the text of each section
    print("\n--- SECTION CONTENTS ---\n")
    for idx, section in enumerate(sections, 1):
        section_text = " ".join(b[4].strip() for b in section)
        print(f"[Section {idx}]\n{section_text}\n")

    # Visualize sections
    fig, ax = plt.subplots(figsize=(8, 11))
    ax.set_xlim(0, page.rect.width)
    ax.set_ylim(page.rect.height, 0)  # invert y-axis

    for i, section in enumerate(sections):
        x0 = min(b[0] for b in section)
        y0 = min(b[1] for b in section)
        x1 = max(b[2] for b in section)
        y1 = max(b[3] for b in section)

        rect = plt.Rectangle((x0, y0), x1 - x0, y1 - y0, edgecolor='red', facecolor='none', lw=2)
        ax.add_patch(rect)
        ax.text(x0, y0 - 5, f"Section {i+1}", color='blue', fontsize=8)

    ax.set_title("Section Detection via PyMuPDF")
    plt.tight_layout()
    plt.show()

# Example usage:
extract_and_visualize_sections("Mayukha Resume.pdf")
```

```python
from pdf2image import convert_from_path
import pytesseract
from pytesseract import Output
import matplotlib.pyplot as plt

import os

# Set path if needed
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def ocr_and_group_sections(pdf_path, spacing_threshold=25):
    # Convert first page of PDF to image
    images = convert_from_path(pdf_path, dpi=300)
    image = images[0]  # First page only

    # Perform OCR with bounding boxes
    data = pytesseract.image_to_data(image, output_type=Output.DICT)

    n_boxes = len(data['level'])
    boxes = []
    for i in range(n_boxes):
        if data['text'][i].strip():
            x, y, w, h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
            text = data['text'][i]
            boxes.append({'x': x, 'y': y, 'w': w, 'h': h, 'text': text})

    # Sort by vertical position (top to bottom)
    boxes.sort(key=lambda b: b['y'])

    # Group into sections based on vertical spacing
    sections = []
    current_section = []
    prev_y = None

    for b in boxes:
        if prev_y is not None and abs(b['y'] - prev_y) > spacing_threshold:
            if current_section:
                sections.append(current_section)
                current_section = []
        current_section.append(b)
        prev_y = b['y']

    if current_section:
        sections.append(current_section)

    # Print each section's text
    print("\n--- OCR SECTION CONTENTS ---\n")
    for i, section in enumerate(sections, 1):
        section_text = " ".join(b['text'] for b in section)
        print(f"[Section {i}]\n{section_text}\n")

    # Visualize
    fig, ax = plt.subplots(figsize=(10, 13))
    ax.imshow(image)

    for i, section in enumerate(sections):
        x0 = min(b['x'] for b in section)
        y0 = min(b['y'] for b in section)
        x1 = max(b['x'] + b['w'] for b in section)
        y1 = max(b['y'] + b['h'] for b in section)

        rect = plt.Rectangle((x0, y0), x1 - x0, y1 - y0, edgecolor='red', facecolor='none', lw=2)
        ax.add_patch(rect)
        ax.text(x0, y0 - 5, f"Section {i+1}", color='blue', fontsize=8)

    plt.axis('off')
    plt.tight_layout()
    plt.show()

# Example usage:
ocr_and_group_sections("Mayukha Resume.pdf")
```

```python
from pdf2image import convert_from_path
import pytesseract
from pytesseract import Output
import matplotlib.pyplot as plt

def ocr_and_group_sections2(pdf_path, spacing_threshold=25, short_word_length=2, x_axis_margin=15):
    # Convert first page of PDF to image
    images = convert_from_path(pdf_path, dpi=300)
    image = images[0]  # First page only

    # OCR
    data = pytesseract.image_to_data(image, output_type=Output.DICT)

    n_boxes = len(data['level'])
    boxes = []
    for i in range(n_boxes):
        if data['text'][i].strip():
            x, y, w, h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
            text = data['text'][i]
            boxes.append({'x': x, 'y': y, 'w': w, 'h': h, 'text': text})

    # Get topmost text
    topmost = min(boxes, key=lambda b: b['y'])
    print(f"Topmost Text: {topmost['text']} (at y={topmost['y']})\n")

    # Sort top to bottom
    boxes.sort(key=lambda b: b['y'])

    # Group into sections by vertical gap
    sections = []
    current_section = []
    prev_y = None

    for b in boxes:
        if prev_y is not None and abs(b['y'] - prev_y) > spacing_threshold:
            if current_section:
                sections.append(current_section)
                current_section = []
        current_section.append(b)
        prev_y = b['y']

    if current_section:
        sections.append(current_section)

    # Merge lonely short-text blocks on same x-axis
    merged_sections = []
    for sec in sections:
        new_sec = []
        carry = None
        for b in sec:
            if len(b['text']) <= short_word_length:
                if carry is None:
                    carry = b
                else:
                    new_sec.append(carry)
                    carry = b
            else:
                if carry:
                    # Try merging carry with closest element in new_sec (based on x proximity)
                    merged = False
                    for s in reversed(new_sec):
                        if abs(s['x'] - carry['x']) <= x_axis_margin:
                            s['text'] += f" {carry['text']}"
                            merged = True
                            break
                    if not merged:
                        new_sec.append(carry)
                    carry = None
                new_sec.append(b)
        if carry:
            new_sec.append(carry)
        merged_sections.append(new_sec)

    # Print each section
    print("\n--- CLEANED OCR SECTION CONTENTS ---\n")
    for i, section in enumerate(merged_sections, 1):
        section_text = " ".join(b['text'] for b in section)
        print(f"[Section {i}]\n{section_text}\n")

    # Visualize
    fig, ax = plt.subplots(figsize=(10, 13))
    ax.imshow(image)

    for i, section in enumerate(merged_sections):
        x0 = min(b['x'] for b in section)
        y0 = min(b['y'] for b in section)
        x1 = max(b['x'] + b['w'] for b in section)
        y1 = max(b['y'] + b['h'] for b in section)

        rect = plt.Rectangle((x0, y0), x1 - x0, y1 - y0,
                             edgecolor='red', facecolor='none', lw=2)
        ax.add_patch(rect)
        ax.text(x0, y0 - 5, f"Section {i+1}", color='blue', fontsize=8)

    # Highlight the topmost box
    ax.add_patch(
        plt.Rectangle((topmost['x'], topmost['y']), topmost['w'], topmost['h'],
                      edgecolor='green', facecolor='none', lw=2, linestyle='--')
    )
    ax.text(topmost['x'], topmost['y'] - 10, 'Topmost Text', color='green', fontsize=8)

    plt.axis('off')
    plt.tight_layout()
    plt.show()

ocr_and_group_sections2("Mayukha Resume.pdf")
```

```python
import fitz  # PyMuPDF
import matplotlib.pyplot as plt

def extract_sections_with_fitz2(pdf_path, spacing_threshold=20, x_axis_threshold=30, min_text_len_to_merge=5):
    doc = fitz.open(pdf_path)
    page = doc[0]  # First page

    # Extract all text blocks [(x0, y0, x1, y1, "text", block_no, block_type)]
    blocks = page.get_text("blocks")
    blocks = [b for b in blocks if b[4].strip()]  # remove empty text blocks

    # Sort blocks by y0 (top-down in PDF space)
    blocks.sort(key=lambda b: b[1])

    # Save topmost data
    topmost_block = min(blocks, key=lambda b: b[1])
    print("\n--- TOPMOST BLOCK ---\n")
    print(topmost_block[4].strip())

    # Group blocks into sections based on vertical spacing
    sections = []
    current_section = []
    prev_y1 = None

    for block in blocks:
        x0, y0, x1, y1, text, *_ = block
        if prev_y1 is not None and (y0 - prev_y1) > spacing_threshold:
            if current_section:
                sections.append(current_section)
                current_section = []
        current_section.append(block)
        prev_y1 = y1

    if current_section:
        sections.append(current_section)

    # Combine "small" sections that are horizontally aligned and too short
    merged_sections = []
    skip_indices = set()

    for i, section in enumerate(sections):
        if i in skip_indices:
            continue

        section_text = " ".join(b[4].strip() for b in section)
        if len(section_text.strip()) > min_text_len_to_merge:
            merged_sections.append(section)
            continue

        # Try merging with the next horizontally-aligned section
        merged = False
        for j in range(i+1, len(sections)):
            if j in skip_indices:
                continue
            next_section = sections[j]
            # Compare horizontal (x-axis) alignment
            x_diff = abs(min(b[0] for b in section) - min(b[0] for b in next_section))
            if x_diff <= x_axis_threshold:
                # Merge sections
                merged_sections.append(section + next_section)
                skip_indices.add(j)
                merged = True
                break
        if not merged:
            merged_sections.append(section)

    # Print section contents
    print("\n--- MERGED SECTION CONTENTS ---\n")
    for idx, section in enumerate(merged_sections, 1):
        section_text = " ".join(b[4].strip() for b in section)
        print(f"[Section {idx}]\n{section_text}\n")

    # Visualize sections
    fig, ax = plt.subplots(figsize=(8, 11))
    ax.set_xlim(0, page.rect.width)
    ax.set_ylim(page.rect.height, 0)  # invert y-axis

    for i, section in enumerate(merged_sections):
        x0 = min(b[0] for b in section)
        y0 = min(b[1] for b in section)
        x1 = max(b[2] for b in section)
        y1 = max(b[3] for b in section)

        rect = plt.Rectangle((x0, y0), x1 - x0, y1 - y0, edgecolor='red', facecolor='none', lw=2)
        ax.add_patch(rect)
        ax.text(x0, y0 - 5, f"Section {i+1}", color='blue', fontsize=8)

    ax.set_title("Improved Section Detection")
    plt.tight_layout()
    plt.show()

# Example usage:
extract_sections_with_fitz2("resumevk.pdf")
```
